### Store JSON data into a DataFrame

```python
json_file = "https://api.covidtracking.com/v1/us/daily.json"
cases = pd.read_json(json_file)
cases.head()
```

### Clean DataFrame

```python
cases.columns
```

```python
cases_df = cases[["date", "positive", "negative"]].copy()
cases_df.head()
```

```python
cases_df.info()
```

```python
clean_cases = cases_df.dropna()
```

### Connect to local database

```python
clean_cases
```

```python
df = clean_cases['date'].astype('str')
```

```python
clean_cases['date'] = df.map(lambda x: dt.datetime.strptime(x,'%Y%m%d'))
```

```python
clean_cases[["positive","negative"]] = clean_cases[["positive","negative"]].astype('int64')
clean_cases.info()
```

```python
clean_cases.sort_values(by=['date'], ascending=True)
```

### Check for tables

```python
clean_cases = clean_cases.loc[(clean_cases["date"] <= '2021-03-15') & (clean_cases["date"] >= '2020-12-21')]
clean_cases
```

### Create engine for Postgres database to connect with pandas

```python
# Create an engine that can talk to the database
engine = create_engine('postgresql+psycopg2://postgres:xxxx@localhost:5432/Covid19_Vaccination_db')
connection = engine.connect();
```

### Load data into postgres

```python
engine.execute("DROP TABLE IF EXISTS vaccine_csv")
clean_cases.to_sql('vaccine_csv', connection)
engine.execute("SELECT * FROM vaccine_csv").fetchall()
```
